```python
def clean_text(text):
    cleaned_text = re.sub('[a-z]', '', text)
    cleaned_text = re.sub('[\{\}\[\]\/?,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          ' ', cleaned_text)
    cleaned_text = re.sub('\n', '', cleaned_text)
    cleaned_text = re.sub('\t', '', cleaned_text)
    cleaned_text = re.sub('캡처', '', cleaned_text)
    return cleaned_text
```

Naver 사회뉴스 조회수 top 30 4월

```python
for apr in tqdm(april):
    url = 'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=102&date=201904' + apr
    driver.get(url)
    response = urllib.request.urlopen(url)
    soup = BeautifulSoup(response, 'lxml')
    for i in range(30):
        title = soup.find_all('div', {"class": "ranking_text"})[i].find('a')['title']
        if title in article_titles:
            continue
        article_titles.append(title)
        new_url ='https://news.naver.com' + soup.find_all('div', {"class": "ranking_text"})[i].find('a')['href']

        article_urls.append(new_url)
        response = urllib.request.urlopen(new_url)
        soup_article = BeautifulSoup(response, 'lxml')
        content = clean_text(soup_article.select_one("#articleBodyContents").text)
        article_contents.append(content)
```

```python
df = pd.DataFrame([article_urls, article_titles, article_contents])
df = df.T
df.columns = ['URL', '제목', '내용']
df.head()
df.nunique()
```

```python
f = lambda x:re.sub('오류를 우회하기 위한 함수 추가', '', x)
f1 = lambda x:re.sub('flash function  flash removeallback', '', x)
f2 = lambda x:re.sub('       C     ', '', x)
df['내용'] = df['내용'].apply(f)
df['내용'] = df['내용'].apply(f1)
df['내용'] = df['내용'].apply(f2)
```
